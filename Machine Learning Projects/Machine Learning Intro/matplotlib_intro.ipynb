### matplotlib example workflow 

```python
# 0. import matplotlib and get it ready for plotting in jupyter 
%matplotlib inline
import matplotlib.pyplot as plt 

# 1. prepare data 
x = [1,2,3,4,5]
y = [10 ,20, 30, 40, 50]

# 2. setup plot 
fig, ax = plt.subplots(figsize = (2,2))

# 3. plot data 
ax.plot(x,y);

# 4. Customize Plot  
ax.set(title = "Sample Plot", 
      xlabel="x-axis", 
      ylabel="y-axis") 

# 5. Save and show (you can sve the whole figure)
fig.savefig("./images/sample-plot.png")
```

## two option for subplots

```python
# subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize= (10,5))
# plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(prices.keys(), prices.values());
ax4.hist(np.random.randn(100));
```

```python
# subplots option 2 
fig, ax = plt.subplots(nrows=2, ncols=2, figsize = (10,5))
# plot to different indexes
ax[0,0].plot(x, x*5);
ax[1,1].scatter(np.random.random(10), np.random.random(10));
```

## which one should be used pyplot or OO
* when plotting something simple use pyplot
* when plotting something complex or feature rich use OO

```python
over_50 = heart_desiese[heart_desiese.age > 50]
over_50.head()
```

```python
over_50.plot(kind = "scatter", 
            x = "age", 
            y = "chol",
            xlabel = "age",
            c = "target"); #c = color 
```

```python
# using OO method 
fig, ax = plt.subplots(figsize = (10,6))
over_50.plot(kind = "scatter", 
            x = "age", 
            y = "chol",
            c = "target",
            ax = ax)
ax.set(xlim=(40,100));
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize = (10,6))

# plot the data 
scatter = ax.scatter(x = over_50.age, 
                    y = over_50.chol, 
                    c = over_50.target)

# customize the plot 
ax.set(title = "Heart Disease Cholestrol level based on age", 
      xlabel = "Age", 
      ylabel = "Cholestrol") 

# add a legend  
ax.legend(*scatter.legend_elements(), title = "target") 

# add horizontal line 
ax.axhline(over_50.chol.mean(), linestyle = "--", c = "red");
```

```python
# changing the style  
plt.style.use("seaborn-v0_8-whitegrid")

# subplot between age chol and thalech 
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize = (7,10), 
                               sharex=True) 

# plotting the data  
ax1.scatter(x = over_50.age, 
           y = over_50.chol,
            cmap="winter",
           c = over_50.target)

ax2.scatter(x = over_50.age, 
           y = over_50.thalach, 
           c = over_50.target, 
           cmap="winter") # chages color of plot

# customizing  
ax1.set( 
       ylabel= "Cholestrol")

ax2.set(xlabel = "Age", 
       ylabel= "Thalach")

# setting legends  
ax1.legend(*scatter.legend_elements(), title = "target")
ax2.legend(*scatter.legend_elements(), title = "target")
# adding horizontal line  
ax1.axhline(y = over_50.chol[over_50.target == 1].mean(), linestyle = "--", c = "Red")
ax2.axhline(y = over_50.thalach[over_50.target == 1].mean(), linestyle = "--", c = "Red")

# add title to the figure  
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");
```
